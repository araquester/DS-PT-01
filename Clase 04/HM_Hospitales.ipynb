```python
def obtener_coordenadas (punto):
    punto = punto.replace("POINT (","")
    punto = punto.replace(")","")
    punto = punto.split()
    longitude = float(punto[0])
    latitude = float(punto[1])
    return longitude, latitude
```

```python
# para no romper el dataframe creo una copia
data_copia = data.copy()
```

```python
data_copia.head()
```

```python
data_copia["longitude"] = data_copia["WKT"].apply(lambda x: obtener_coordenadas(x)[0])
```

```python
data_copia.head()
```

```python
data_copia["latitude"] = data_copia["WKT"].apply(lambda x: obtener_coordenadas(x)[1])
```
